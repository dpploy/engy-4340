```python
import math
import numpy as np
```

```python
def l_dist(time, k_over_mu, delta_x_0, delta_v_0):
    
    aa = delta_v_0 / math.sqrt(k_over_mu)
    bb = delta_x_0
    dist = aa * np.sin(math.sqrt(k_over_mu) * time) + bb * np.cos(math.sqrt(k_over_mu) * time)
    return dist
```

```python
x1_0 = 1
v1_0 = -2
m1 = 1

x2_0 = 3
v2_0 = 5
m2 = 1.5

delta_v_0 = v2_0-v1_0
l_0 = x2_0 - x1_0

k = 2
mu = m1 * m2 / (m1 + m2)

f = 1/(2*math.pi)*math.sqrt(k/mu)
T = 1/f

times = np.linspace(0,3*T,200)
l_vals = l_dist(times, k/mu, x2_0-x1_0, delta_v_0)

amplitude = (l_0**2*math.sqrt(k/mu) + delta_v_0**2/math.sqrt(k/mu))/math.sqrt(l_0**2 * k/mu + delta_v_0**2)
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline
#plt.style.use('dark_background')
plt.style.use('classic')

plt.figure(1, figsize=(12, 7))

plt.plot(times, l_vals/amplitude, 'r*', label='l_minus', markersize=5)
plt.grid(True)
plt.show()
```
