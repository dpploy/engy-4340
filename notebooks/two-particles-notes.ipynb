```python
import math
import numpy as np
```

```python
def l_plus(time, k_mu, delta_x_0, delta_v_0):
    
    aa = delta_v_0 / math.sqrt(k_mu)
    bb = delta_x_0
    
    l_plus = aa * math.sinh(math.sqrt(k_mu) * time) + bb * math.cosh(math.sqrt(k_mu) * time)
    return l_plus if l_plus > 0.0 else 0.0
```

```python
def l_minus(time, k_mu, delta_x_0, delta_v_0):
    
    aa = delta_v_0 / math.sqrt(k_mu)
    bb = delta_x_0
    l_minus = aa * math.sin(math.sqrt(k_mu) * time) + bb * math.cos(math.sqrt(k_mu) * time)
    return l_minus
```

```python
x1_0 = 1
v1_0 = 0
m1 = 1

x2_0 = 2
v2_0 = 0
m2 = 1.5

k = 2
mu = m1 + m2 / (m1 * m2)

times = np.linspace(0,10,1000)
l_minus_vals = [l_minus(time, k*mu, x2_0-x1_0, v2_0-v1_0) for time in times]
l_plus_vals = [l_plus(time, k*mu, x2_0-x1_0, v2_0-v1_0) for time in times]
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline
#plt.style.use('dark_background')
plt.style.use('classic')

plt.figure(1, figsize=(12, 7))

#plt.plot(times, l_plus_vals, 'b*', label='l_plus', markersize=10)
plt.plot(times, l_minus_vals, 'r*', label='l_minus', markersize=10)
plt.grid(True)
plt.show()
```
